```python
all_data2 = pd.read_csv('all_data2.csv')
Hansen_data = pd.read_csv('89_solvents.csv')


names = all_data2['Solvent'].values[57:]

for s in names:
    try:
        upperstring = s.upper()
        row = Hansen_data[Hansen_data['Name'].upper()== upperstring]
        all_data2.loc[all_data2.Solvent == s, 'Dispersion'] = row['Hansen Delta D (MPa1/2)']
        all_data2.loc[all_data2.Solvent == s, 'Polarity'] = row['Hansen Delta P (MPa1/2)']
        all_data2.loc[all_data2.Solvent == s, 'H Bonding'] = row['Hansen Delta H (MPa1/2)']
    except KeyError:
        print(s)
    
    
all_data2.to_csv('all_data3.csv')

```

```python
Hansen_data = pd.read_csv('h_s_data.csv')
hsp_data = pd.DataFrame()

pattern = r"([A-Z]-(?=[A-Z]))|([a-z](?=[0-9])|[0-9](?=[A-Z]))"

for index, row in Hansen_data.iterrows():
    name = row['Solvent'].split('"')[-2]
    volume = row['Volume'].split('[')[1].split(',')[0]
    h_d = row['h_d'].split('[')[1].split(',')[0]
    h_p = row['h_p'].split('[')[1].split(',')[0]
    h_h = row['h_h'].split('[')[1].split(',')[0]
    print(volume)
    print(h_d)
    print(h_p)
    print(h_h)
    result = re.sub(
        pattern,
        lambda x: x.group(1)[:-1] if x.group(1) else x.group(2).capitalize() + "-",
        name,
    )
    #print(result)
    hsp_dict = pd.DataFrame({'Solvent': result,
                'Volume': volume,
                'h_d': h_d,
                'h_p': h_p,
                'h_h': h_h,
                }, index = [0])
    hsp_data = pd.concat([hsp_data, hsp_dict])


hsp_data.to_csv('hsp_data_processed.csv')
    

    





```
